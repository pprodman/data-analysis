```python
summary = []

for column in df.columns:
    col_data = df[column]

    # Informacin bsica
    total_rows = len(col_data)
    non_null_count = col_data.notna().sum()
    null_count = total_rows - non_null_count
    percent_complete = (non_null_count / total_rows) * 100 if total_rows > 0 else 0

    if pd.api.types.is_numeric_dtype(col_data):
        col_clean = pd.to_numeric(col_data, errors='coerce')

        # Separar parte entera y decimal
        integer_part = col_clean.dropna().abs().apply(lambda x: int(x))
        decimal_part = col_clean.dropna().apply(lambda x: str(x).split('.')[1] if '.' in str(x) and len(str(x).split('.')[1]) > 0 else '')

        # Contar dgitos
        max_int_digits = integer_part.apply(lambda x: len(str(x)) if x != 0 else 1).max() if not integer_part.empty else np.nan
        max_dec_digits = decimal_part.apply(len).max() if not decimal_part.empty else 0

        summary.append({
            'column': column,
            'type': 'numeric',
            'dtype': str(col_data.dtype),
            'null_count': null_count,
            'percent_complete': round(percent_complete, 2),
            'max_integer_digits': max_int_digits,
            'max_decimal_digits': max_dec_digits
        })

    else:
        try:
            max_length = col_data.astype(str).str.len().max()
        except:
            max_length = np.nan

        summary.append({
            'column': column,
            'type': 'text',
            'dtype': str(col_data.dtype),
            'null_count': null_count,
            'percent_complete': round(percent_complete, 2),
            'max_length': max_length
        })

# Convertir a DataFrame
summary_df = pd.DataFrame(summary)

summary_df
```

```python
print(f"En la tabla {table} existen las siguientes columnas de tipo numrico:")

for column in summary:
    if column['type'] == 'numeric':
        col_name = column.get('column')  # o el nombre correcto segn summary
        print(f"--> Columna '{col_name}':")

        # Convertir a numrico, forzando errores a NaN
        numeric_col = pd.to_numeric(df[col_name], errors='coerce')

        # Verificar si hay decimales distintos de .0
        has_non_zero_decimal = ((numeric_col % 1) != 0).any()

        if has_non_zero_decimal:
            print("   Existen decimales diferentes a .0")
        else:
            print("   Todos los valores tienen decimales iguales a .0")
            df[col_name] = numeric_col.astype('Int64')
            print("   Se han eliminado los decimales .0 y convertido a entero.")
```

```python
from itertools import combinations

found_pk = False  # Inicialmente no se ha encontrado clave primaria

# 1. Primero, intentamos encontrar una columna nica
for column in df.columns:
    if df[column].is_unique:
        print(f"La columna '{column}' es nica y podra ser una clave primaria.")
        found_pk = True
        break  # Salir inmediatamente si se encuentra

# 2. Si no hay columna nica, buscamos combinaciones
if not found_pk:
    columns = df.columns
    for r in range(2, len(columns) + 1):
        for combo in combinations(columns, r):
            if df.duplicated(subset=combo).sum() == 0:
                print(f"La combinacin de columnas {combo} podra ser una clave primaria compuesta.")
                found_pk = True
                break  # Rompe el bucle de combinaciones
        if found_pk:
            break  # Rompe el bucle de tamaos
```
